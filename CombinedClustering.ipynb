## Preprocessing the data


```python
file_path = "/content/drive/MyDrive/data-mining/Wine.csv" # wine dataset file path
df = pd.read_csv(file_path,  sep='\t', on_bad_lines='skip') # load and read dataset
```

```python
cP=ClusterProcess(n_pca_components=3)
```

```python
df[['Alcohol', 'Malic_Acid','Ash','Ash_Alcanity','Magnesium','Total_Phenols','Flavanoids','Nonflavanoid_Phenols','Proanthocyanins','Color_Intensity','Hue','OD280','Proline']] = df["Alcohol,Malic_Acid,Ash,Ash_Alcanity,Magnesium,Total_Phenols,Flavanoids,Nonflavanoid_Phenols,Proanthocyanins,Color_Intensity,Hue,OD280,Proline"].str.split(",", 13, expand = True)
del df[df.columns[0]]
df = df.astype(float)
print(df.dtypes)
```

### Performing Principal Component Analysis

```python
pca_data=cP.findPCA(df)
pca_data
```

### Visualizing 3D Scatter Plot

```python
cP.plot3dScatter(pca_data)
```

## **Agglomerative Clustering**

### Computing Silhouette Score

```python
s_score=[]
for i in range(2,11):
    predicts,labels,score_s=cP.do_clustering(pca_data,i,cluster.AgglomerativeClustering)
    s_score.append(score_s)
s_score=[[ i+2,idx ] for i,idx in enumerate(s_score)]
```

```python
cP.plotBar(s_score,['cluster_size','silhouette'], color='silhouette')
```

```python
predicts,labels,score_s=cP.do_clustering(pca_data,3,cluster.AgglomerativeClustering)
```

```python
pca_data['cluster']=predicts
```

```python
cP.plot3dScatter(pca_data,color='cluster')
```

```python
df['cluster']=predicts
```

## **K Means Clustering**

### Computing Silhouette Score as well as Inertia Score

```python
s_score=[]
inr_score=[]
for i in range(2,12):
    predicts,clusters_,labels,score_s,score_inr=cP.do_clustering(pca_data,i,cluster.KMeans)
    s_score.append(score_s)
    inr_score.append(score_inr)
s_score=[[ i+2,idx ] for i,idx in enumerate(s_score)]
inr_score=[[ i+2,idx ] for i,idx in enumerate(inr_score)]
```

```python
cP.plotBar(inr_score,['cluster_size','silhouette'],color='silhouette')
```

```python
cP.plotBar(s_score,['cluster_size','silhouette'],color='silhouette')
```
